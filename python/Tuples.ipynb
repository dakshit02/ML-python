```python
#Tuple Operations

#concatenation
tup1= (1,2,3,4)
tup2 = ("hello",3.14,True)
concatenated_tuple = tup1+tup2
print(concatenated_tuple)
```

```python
print(tup2 *3 )  #multiplying the tuple with a number will multiply it by the number given not inside but in the form of occurence
print(tup1 *5 )   
```

```python
#immutability of tuple 
tuple_1 = ("dakshit",3,2,True)
tuple_1[0]="chandel"
print(tuple_1)
```

```python
#Tuples Methods
nums = (10,22,33,40,50,10)
print(nums.count(10))
print(nums.index(22))
print(nums.index(10))
```

```python
#packing and unpacking the tuples

#packing a tuple 
packed_tuples= 1,"hello",3.14
print(packed_tuples)
```

```python
#unpacking a tuple
a,b,c = packed_tuples
print(a,b,c)
```

```python
packed_tup = 1,2,3,4,5,6
first ,*middle, last = packed_tup
print(first)
print(middle)
print(last)
```
