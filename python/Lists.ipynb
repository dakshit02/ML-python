```python
##accessing list elements
fruits=["Apple","banana","cherry"]
print(fruits[0])
print(fruits[1])
print(fruits[2])
print(fruits[-1])    #means get the last element
```

```python
print(fruits[0:])
print(fruits[0:2]) #excluding 2 will print the rest of the items
print(fruits[1:])
print(fruits[-1:-3])
```

```python
#modifying the list elements 
fruits[1]="watermelon"     #changed bananan from watermelon
print(fruits)
```

```python
#List Methods
fruits.append("orange")  #add an item at the end
print(fruits)
```

```python
fruits.insert(1,"dakshit")
```

```python
print(fruits)
```

```python
fruits.remove("orange")
```

```python
#remove and return the last item 
poped_item = fruits.pop()
print(poped_item)
```
