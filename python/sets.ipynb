```python
##Basic Set operation
#add and remove elements 

new_set=set([45,67,78])
new_set.add(6)
print(new_set)
```

```python
#remove
new_set_1 = set([1,2,8,7])
new_set_1.remove(8)
print(new_set_1) 
```

```python
# using remove when the element is not present in the set will throw error
new_set_1.remove(11)
print(new_set_1)
```

```python
#so instead we can use discard () it won't throw error if the element is not present in the set 
new_set_1.discard(11)
print(new_set_1)
```

```python
#pop
removed_element = new_set.pop()
print(removed_element)
print(new_set)

  
```

```python
#clear 
my_set.clear()
print(my_set)
```

```python
#set membership test 
set_test= set([12,13,14,15])
print(3 in set_test)
print(13 in set_test)

```

```python
#mathematical operations
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

#union
union_set= set1.union(set2)
print(union_set)

#intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

#intersection update - updates the intersected elemnts in the first set 
set1.intersection_update(set2)
print(set1)


 
```

```python
#difference
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
print(set1.difference(set2))
print(set2.difference(set1))
```

```python
##Symmetric Difference 
print(set1.symmetric_difference(set2))    #uniques elemnts from both the elements are combined or intersection elements are removed 

```

```python
#Sets Methods
set1 = {1,2,3}
set2 = {3,4,5}
set3 = {2,3}

#is subset

print(set1.issubset (set2))
print(set3.issubset(set1))
 

```
